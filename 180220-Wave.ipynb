```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_style('white')
sns.set_context("poster")

df = pd.read_hdf('data1-zdc.h5', 'table', where='sensor_id in 113', columns=['val', 'timestamp'])
```

```python
df = df.set_index('timestamp')
df.head()
```

```python
df.tail()
```

```python
plt.plot(df.val)
```

```python
t = df.index
sp = np.fft.fft(df.val)
freq = df.index.shape[-1]
```

```python
type(sp.real)
```

```python
len(sp.real)
```

```python
x = np.arange(0, len(sp.real))
```

```python
sp.real[0:500]
```

```python
# Plot spectrum
plt.figure(dpi=300)
plt.plot(x, np.abs(sp.real))
# plt.plot(x, sp.real)
plt.xlim(0, len(sp.real)/2)
plt.ylim(0, 10e3)
```
