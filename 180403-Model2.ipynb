```python
trf_train = dft.loc['2016-09-01':'2017-01-05'].trf_count
trf_train.isnull().sum()
```

## Decompose the Data

```python
trf_train = trf_train.resample('1800S').median()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

X = trf_train.values
```

```python
decomposition = seasonal_decompose(X, freq=336)
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.subplot(411)
plt.plot(trf_train, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```

What's the problem? 

There's a sharp downward trend at the end of the series

## ARIMA

```python
from statsmodels.tsa.stattools import acf, pacf
```

```python
# ACF and PACF plots:
lags = 48
lag_acf = acf(X, nlags=lags)
lag_pacf = pacf(X, nlags=lags, method='ols')
```

```python
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf, '.')
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(X)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(X)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf, '.')
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(X)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(X)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

p = 5, q = 9

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model = ARIMA(X, order=(9, 0, 5)).fit()
```

13.3807

```python
## Combined Model
plt.figure()
plt.plot(X)
plt.plot(model.fittedvalues, 'r--', alpha=0.5)
plt.title('MSE: {:.4f}'.format(sum((model.fittedvalues-X)**2)/len(X)))
```

```python
y1 = X
start_date = pd.to_datetime('2017-01-01')
end_date = pd.to_datetime('2017-03-01')
y2 = model.predict()
x1 = np.arange(len(y1))
x2 = np.arange(len(y2)) + x1[-1]
```
