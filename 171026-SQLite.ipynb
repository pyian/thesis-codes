```python
# Import zero-offset values from SQLite to pandas

%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import sqlite3

conn = sqlite3.connect("db_description.sqlite")
```

```python
cur = conn.cursor()
```

```python
cur.execute("select * from system_config limit 5;")

results = cur.fetchall()
print(results)
```

```python
command = "SELECT sc.sensor_id, sc.zero_offset_value, cl.start_timestamp\
 FROM sensor_config sc\
 INNER JOIN system_config sysc ON sc.sensor_id = sysc.sensor_id AND sc.sensor_config_id = sysc.sensor_config_id\
 INNER JOIN config_log cl ON sysc.active_system_config_id = cl.active_system_config_id\
 WHERE sc.sensor_id = 64 AND cl.start_timestamp <= '2017-06-01 08:00:00' AND (cl.end_timestamp >= '2015-06-01 08:00:00' OR cl.end_timestamp IS NULL);"
```

```python
cur.execute(command)
results = cur.fetchall()
print(results)
```

## Using Pandas
